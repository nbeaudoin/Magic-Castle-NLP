# Goal: 
Create a dataframe that has all review text, date and the rating

Helpful sources:
 - https://www.scrapingdog.com/blog/how-to-scrape-yelp-reviews
 - https://www.nicholasrenotte.com/how-to-build-a-sentiment-analyser-for-yelp-reviews-in-python/

```python
from bs4 import BeautifulSoup
import requests
import nltk 
import re
import pandas as pd
import time
```

# Download HTML with BeautifulSoup

```python
# Get raw HTML
r = requests.get('https://www.yelp.com/biz/the-magic-castle-hollywood').text
```

```python
soup = BeautifulSoup(r, 'html.parser')
```

```python
allrev = soup.find_all("li",{"class": "margin-b5__373c0__2ErL8 border-color--default__373c0__2oFDT"})
```

```python
dates = []
reviews = []
ratings = []
```

## Scrape one page

```python
for i in range(1, len(allrev)):
    ### DATE ###
    dates.append(allrev[i].find("span", {"class": "css-e81eai"}).text)
    print(allrev[i].find("span", {"class": "css-e81eai"}).text)
    
    ### REVIEW ###
    reviews.append(allrev[i].find("p", {"class": "comment__373c0__1M-px css-n6i4z7"}, {"class": "raw__373c0__3rcx7"}).get_text())
    print(allrev[i].find("p", {"class": "comment__373c0__1M-px css-n6i4z7"}, {"class": "raw__373c0__3rcx7"}).get_text())
    
    ### RATING ###
    ratings.append(allrev[i].findAll("span", {"class": "display--inline__373c0__2SfH_ border-color--default__373c0__30oMI"})[0].div["aria-label"])
    print(allrev[i].findAll("span", {"class": "display--inline__373c0__2SfH_ border-color--default__373c0__30oMI"})[0].div["aria-label"])
    
    print('   ')
```

## Convert lists to Pandas DF

```python
df = pd.DataFrame(
    {'Date': dates,
     'Review': reviews,
     'Rating':ratings
    })

df.head()
```

## Scrape multiple pages

```python
dates = []
reviews = []
ratings = []
```

```python
for page in range(10, 2410, 10):
    
    url = 'https://www.yelp.com/biz/the-magic-castle-hollywood?start={}'.format(page)
    print(url)
    
    # Get raw HTML
    r = requests.get(url).text
    soup = BeautifulSoup(r, 'html.parser')
    allrev = soup.find_all("li",{"class": "margin-b5__373c0__2ErL8 border-color--default__373c0__2oFDT"})
    
    print('Scraping...')
    
    counter = 0
    
    for i in range(1, len(allrev)):
         
        ### DATE ###
        dates.append(allrev[i].find("span", {"class": "css-e81eai"}).text)
        #print(allrev[i].find("span", {"class": "css-e81eai"}).text)

        ### REVIEW ###
        reviews.append(allrev[i].find("p", {"class": "comment__373c0__1M-px css-n6i4z7"}, {"class": "raw__373c0__3rcx7"}).get_text())
        #print(allrev[i].find("p", {"class": "comment__373c0__1M-px css-n6i4z7"}, {"class": "raw__373c0__3rcx7"}).get_text())

        ### RATING ###
        ratings.append(allrev[i].findAll("span", {"class": "display--inline__373c0__2SfH_ border-color--default__373c0__30oMI"})[0].div["aria-label"])
        #print(allrev[i].findAll("span", {"class": "display--inline__373c0__2SfH_ border-color--default__373c0__30oMI"})[0].div["aria-label"])

        counter += 10
        
        print('{}% complete'.format(counter))
        
        time.sleep(3)
        
```

```python
df = pd.DataFrame(
    {'Date': dates,
     'Review': reviews,
     'Rating':ratings
    })

df.head()
```
